# Exploratory Data Analysis 
## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
attacks_by_country = df.groupby('Country')['Year'].count().reset_index().sort_values('Year', ascending=False)
attacks_by_region = df.groupby('rgn')['Year'].count().reset_index().sort_values('Year', ascending=False)
```

```python
plt.figure(figsize=(16, 6))
plt.title('Number of terrorist attacks by country')
sns.barplot(x='Year', y='Country', data=attacks_by_country.head(20), palette='tab10')
plt.xlabel('Number of attacks')
plt.ylabel('')
plt.show()

plt.figure(figsize=(10, 6))
plt.title('Number of terrorist attacks by region')
sns.barplot(x='Year', y='rgn', data=attacks_by_region, palette='tab10')
plt.xlabel('Number of attacks')
plt.ylabel('')
plt.show()
```
